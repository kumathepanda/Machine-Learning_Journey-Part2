```python
import pandas as pd
import numpy as np
```

```python
data = {'R&D Spend':[165.0,101.0,29.0],'Administration':[137.0,92.0,127.0],'Marketing Spend':[470.0,250.0,201.0],'Profit':[192.0,144.0,91.0]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['f0(x)']=df['Profit'].values.mean()
```

```python
df
```

```python
df['ri1'] = df['Profit'] -  df['f0(x)']
```

```python
df
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt1 = DecisionTreeRegressor(max_depth=1)
```

```python
dt1.fit(df.iloc[:,0:3].values,df['ri1'].values)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt1)
```

```python
df
```

```python
df['preds1'] = df['f0(x)'] + 0.1*df['ri1']
```

```python
df
```

```python
dt2 = DecisionTreeRegressor(max_depth=1)
```

```python
df['ri2'] = df['Profit'] - df['preds1']
```

```python
dt2.fit(df.iloc[:,0:3].values,df['ri2'].values)
```

```python
plot_tree(dt2)
```
