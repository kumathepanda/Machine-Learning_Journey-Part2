```python
def capping_func(X):
    X_df = pd.DataFrame(X, columns=['Area', 'Production', 'Annual_Rainfall', 'Fertilizer', 'Pesticide'])  # Use actual column names or indices
    for col in X_df.columns:
        Lower_limit = X_df[col].quantile(0.025)
        Upper_limit = X_df[col].quantile(0.975)
        X_df[col] = np.where(X_df[col] > Upper_limit, Upper_limit,
                             np.where(X_df[col] < Lower_limit, Lower_limit, X_df[col]))
    return X_df.values
```

```python
trf3 = ColumnTransformer([
    ('trf3',FunctionTransformer(capping_func),[4,5,6,7,8])
],remainder='passthrough')
```

```python
better_pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)
])
```

```python
pipeline = Pipeline(steps=[
    ('better_pipe', better_pipe),  # Add the preprocessing pipeline
    ('regressor', DecisionTreeRegressor())  # Add the estimator as the final step
])
```

```python
pipeline.fit(x_train,y_train)
```
