```python
data = data[ ['price','brokered_by', 'status', 'street',
       'city', 'state', 'zip_code', 'prev_sold_date','bed', 'bath', 'acre_lot','house_size']]
```

```python
columns_to_replace = ['bed', 'bath', 'brokered_by', 'street', 'zip_code']

for col in columns_to_replace:
   
    data[col] = data[col].fillna(0).astype('Int64')  
    
   
    data[col] = data[col].replace(0, pd.NA)
```

```python
def random_imputation(x):
    for i in x.columns:
        x[i][x[i].isnull()] = x[i].dropna().sample(x[i].isnull().sum()).values
```

```python
trf1 = ColumnTransformer([
    ('random_inmputer',FunctionTransformer(random_imputation),slice(0,11))
],remainder='passthrough')
```
